```python
import os
from dotenv import load_dotenv
from langchain_chroma import Chroma
from langchain_openai import OpenAIEmbeddings

load_dotenv()

# Configure paths and parameters
DB_PATH = os.environ.get("CHROMA_DB_PATH", "./chroma_db_cards")
TOP_K = 3
QUERY = "Tell me about the bobbit worm card."

embeddings = OpenAIEmbeddings(api_key=os.environ.get("OPENAI_API_KEY", ""))
chroma = Chroma(persist_directory=DB_PATH, embedding_function=embeddings)
retriever = chroma.as_retriever(search_kwargs={"k": TOP_K})

results = retriever.invoke(QUERY)
print(f"Query: {QUERY}\n")
print("Top Results:")
for i, doc in enumerate(results, start=1):
    print(f"--- Result {i} ---")
    print(doc.page_content)
    if doc.metadata:
        print("metadata:", doc.metadata)
```
